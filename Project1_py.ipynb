#**Business Question**

#1. Does gender affect scores?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

avg_scores = df.groupby("gender")[["math score",
                                   "reading score",
                                   "writing score"]].mean()

avg_scores.plot(kind="bar", figsize=(8,5))

plt.title("Average Subject Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title="Subjects")

plt.show()
```

"This chart compares the average Math, Reading, and Writing scores of male and female students. We can observe that female students scored higher in Reading and Writing, while male students performed slightly better or similarly in Math. Therefore, gender appears to have some influence on student performance in this dataset."

#2. Impact of parental education?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

# Create average score
df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

plt.figure(figsize=(10,5))

sns.barplot(
    x="parental level of education",
    y="average_score",
    data=df,
    estimator='mean'
)

plt.title("Impact of Parental Education on Student Performance")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Score")
plt.xticks(rotation=45)

plt.show()
```

"I used a bar chart because parental education is a categorical variable and student score is a numerical variable. The chart compares average scores across different parental education levels. The results show that students whose parents have higher education generally perform better academically, indicating that parental education has a positive impact on student performance."

#3. Lunch effect on performance ?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

plt.figure(figsize=(6,4))
sns.countplot(x="lunch",hue='lunch',data=df, palette="Set1")

plt.title("Distribution of Students by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Number of Students")

plt.show()
```

"This bar chart compares the average scores of students based on lunch type. Students with standard lunch achieved higher average scores than students with free/reduced lunch, suggesting that lunch type may influence academic performance."

#4. which subject has highest variability?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")

# Calculate standard deviation
std_scores = df[["math score",
                 "reading score",
                 "writing score"]].std()

# Convert to DataFrame
std_df = std_scores.reset_index()
std_df.columns = ["Subject", "Standard Deviation"]

# Create plot
plt.figure(figsize=(7,5))

ax = sns.barplot(
    x="Subject",hue="Subject",
    y="Standard Deviation",
    data=std_df,
    palette="Set2"
)

# Add data annotations

for i, value in enumerate(std_df["Standard Deviation"]):
    ax.text(
        i,                    # x-position
        value + 0.2,          # y-position
        f"{value:.2f}",       # text
        ha='center',
        fontsize=10
    )

plt.title("Variability of Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Standard Deviation")

plt.show()
```

"The subject with the highest standard deviation has the highest variability. This indicates that student performance is less consistent in that subject compared to the others."
